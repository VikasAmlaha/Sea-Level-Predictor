# <B>SEA LEVEL PREDICTOR</B>

# We will analyze a dataset of the global average sea level change since 1880. We will use the data to predict the sea level change through year 2050.

## 1. Used Pandas to import the data from epa-sea-level.csv file.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

df = pd.read_csv('epa-sea-level.csv')
print(df.shape)
df.head()
```

## 2. Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.xlabel('Year', fontsize = 12)
plt.ylabel('CSIRO Adjusted Sea Level', fontsize = 12)
plt.title('Rise in Sea Level', fontsize = 14)
plt.show()
```

## 3. Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python


# Load your data into df (assuming 'Year' and 'CSIRO Adjusted Sea Level' exist)
# df = pd.read_csv('your_data.csv')  # Uncomment this if loading from a file

# Perform linear regression on the entire dataset
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Generate years up to 2050
years = np.arange(df['Year'].min(), 2051)  # Extend to 2050
sea_levels = slope * years + intercept  # Predicted sea levels

# Create scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='blue', alpha=0.7, edgecolors='black', label="Data Points")

# Plot best-fit line
plt.plot(years, sea_levels, color='red', linestyle='-', linewidth=2, label="Best Fit Line (Projection)")

# Labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('CSIRO Adjusted Sea Level (in mm)', fontsize=12)
plt.title('Rise in Sea Level with Projection to 2050', fontsize=14)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

## 4. Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```python


# Load your data (ensure 'Year' and 'CSIRO Adjusted Sea Level' exist)
# df = pd.read_csv('your_data.csv')  # Uncomment this if loading from a file

# Perform linear regression on the full dataset
slope1, intercept1, _, _, _ = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Generate predictions for full dataset trend (from min year to 2050)
years_full = np.arange(df['Year'].min(), 2051)
sea_levels_full = slope1 * years_full + intercept1

# **Step 1: Filter data from 2000 onward**
df_recent = df[df['Year'] >= 2000]

# **Step 2: Perform linear regression on filtered dataset**
slope2, intercept2, _, _, _ = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])

# **Step 3: Generate predictions for recent trend**
years_recent = np.arange(2000, 2051)  # From 2000 to 2050
sea_levels_recent = slope2 * years_recent + intercept2

# **Step 4: Plot everything**
plt.figure(figsize=(10, 5))

# Scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='blue', alpha=0.7, edgecolors='black', label="Data Points")

# First best-fit line (Full dataset)
plt.plot(years_full, sea_levels_full, color='red', linestyle='-', linewidth=2, label="Best Fit (All Data)")

# Second best-fit line (2000 onward)
plt.plot(years_recent, sea_levels_recent, color='green', linestyle='--', linewidth=2, label="Best Fit (2000-Present)")

# Labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('CSIRO Adjusted Sea Level (in mm)', fontsize=12)
plt.title('Rise in Sea Level with Projections to 2050', fontsize=14)
plt.legend()
plt.grid(True)

# **Annotate 2050 predictions**
plt.text(2050, slope1 * 2050 + intercept1, f'All Data: {slope1 * 2050 + intercept1:.2f} mm',
         fontsize=12, color='red')
plt.text(2050, slope2 * 2050 + intercept2, f'2000+: {slope2 * 2050 + intercept2:.2f} mm',
         fontsize=12, color='green')

# Show the plot
plt.show()
```
